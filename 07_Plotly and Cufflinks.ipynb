# Plotly-and-Cufflinks
### Plotly is an interactive vizualization library Cufflinks connects plotly with pandas

```python
import numpy as np
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 
import cufflinks as cf
init_notebook_mode(connected=True)
import plotly.graph_objs as go
cf.go_offline()
```

```python
%matplotlib inline
```

```python
df1 = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df1
```

```python
df2 = pd.DataFrame({'Catagory':['A','B','C'],'Values':[32,42,50]})
df2
```

```python
df1.iplot()
```

```python
df1.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df2.plot(kind='bar',x='Catagory',y='Values')
```

```python
df1.sum().iplot(kind='bar')
```

```python
df1.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[100,200,300,400,500]})
df3
```

```python
df3.iplot(kind='surface')
```

```python
df1.iplot(kind='hist')
```

```python
df1[['A','B']].iplot(kind='spread')
```

```python
df1.iplot(kind='bubble',x='A',y='B',size='C')
```

```python
df1.scatter_matrix()
```
