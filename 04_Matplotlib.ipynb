# Matplotlib

## Plotting

```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```
x
```

```
y
```

```
#Function
plt.plot(x,y)
#if not on JP notebook use below
#plt.show()
```

```
plt.plot(x,y)
plt.xlabel('Xlabel')
plt.ylabel('Ylable')
plt.title('Title')
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
#Object oriented method
fig = plt.figure()
# [left,bottom,width,height]
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Xaxis')
axes.set_ylabel('Yaxis')
axes.set_title('title')
```

```
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("Larger Plot")
axes2.plot(y,x,'r')
axes2.set_title("Smaller Plot")
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title("1st")
axes[1].plot(x,y)
```

## Figure size and dpi

```
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0.1,0,1,0.5])
ax.plot(x,y)
```

```
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(3,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

```
#to save fig
fig.savefig('mygraph.png')
```

```
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])

ax1.plot(x,x**2,label='sq')
ax1.plot(x,x**3,label='cb')
ax1.legend(loc=0)
#loc 0 will choose best positin
```

## Appearence

```
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax1.plot(x,y,color='blue',linewidth=3,alpha=0.5)
#can write lw insted of linewidth
ax1.plot(y,x,color='green',lw=3,ls="--")
#linestyle can be written as ls
ax1.plot(y,x**5,color='red',lw=3,marker='o',markersize=20)
#markeredgecolor markeredgewidth
```

```
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax1.plot(x,y,color='blue',linewidth=3,alpha=0.5)
# to get value between limit
ax1.set_xlim([2,3])
```
