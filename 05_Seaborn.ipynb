# Seaborn
### Statistical plotting library 

### Distribution plots

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
#built in dataset
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
#bins determine size
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#kind = kde
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

### Categorical Plots

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.max)
```

```python
sns.countplot(x='sex',data=tips)
#no need of y it shows count
```

```python
sns.boxenplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxenplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#calling all plot in one line
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

### Matrix plot

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# for heatmap table must be having index
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp = flights.pivot_table(index='year',columns='month',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,linecolor='white',linewidths=2,cmap='coolwarm')
```

```python
sns.clustermap(fp)
```

### Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

### Regression Plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```
